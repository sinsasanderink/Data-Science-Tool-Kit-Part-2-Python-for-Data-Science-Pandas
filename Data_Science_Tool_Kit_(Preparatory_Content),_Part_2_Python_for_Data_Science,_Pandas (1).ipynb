7. Pivoting and Grouping

````python
import pandas as pd

# Create a DataFrame with more details
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'Alice', 'Bob'],
    'Department': ['HR', 'HR', 'IT', 'IT', 'HR'],
    'Salary': [50000, 40000, 70000, 55000, 45000]
}
df = pd.DataFrame(data)

# Grouping by 'Department' to find average salary only
grouped_df = df.groupby('Department')['Salary'].mean()
print(grouped_df)

print("\n ___________ \n")

# Pivoting the table to see salaries and counts of employees per department
pivot_table = pd.pivot_table(df, values='Salary', index='Department', aggfunc=['mean', 'count'])
print(pivot_table)
````

# Pandas - Rows and Columns

Row Indices:
- By default, rows are numbered starting from 0, just like how we count items in a list. This number is the row index.
- For example, in a DataFrame with three rows, the indices would be 0, 1, and 2.

Column Headers:
- Columns also have labels, typically found in the first row of your data file (like CSV). These labels are used as column headers.
- If a data file doesn't have headers, Pandas can automatically assign numerical labels starting from 0, similar to row indices.

Setting Row Indices:
- You can choose one of the DataFrame’s columns to be the row index. For example, if you have a column with unique IDs for each entry, you can set this column as the row index.

Setting Column Headers:
- You can manually assign names to the columns if your data file doesn’t contain any headers or if you want to change them.

````python
import pandas as pd

# Data without headers and default indices
data = [[1, 'Alice', 85],
        [2, 'Bob', 90],
        [3, 'Cathy', 88]]

# Creating a DataFrame without headers
df = pd.DataFrame(data, columns=['ID', 'Name', 'Score'])

print("DataFrame with custom headers and default indices:")
print(df)

# Setting 'ID' as the index
df.set_index('ID', inplace=True)

print("\nDataFrame with 'ID' as the row index:")
print(df)
````

Time-series data refers to a series of data points that are indexed over time. The data is recorded over regular time intervals and is stored along with the time it was recorded. Some common examples of time series include stock prices, temperature, weather report, etc., as this information would make sense only when presented with the time it was recorded.



If a date–time variable has values in the form of a string, then you can call the ‘parse_dates’ function while loading the data into the Pandas DataFrame.
````python
# another way of doing this
import pandas as pd

# Load the dataset
url = 'https://media-doselect.s3.amazonaws.com/generic/NMgEjwkAEGGQZBoNYGr9Ld7w0/rating.csv'
df = pd.read_csv(url)

# Assume Training column is determined by some condition like before
df["Training"] = df['Rating'].apply(lambda x: "Yes" if x < 3.5 else "No")

# List of departments for iteration
departments = ['Finance', 'HR', 'Sales', 'Marketing']

# Initialize a dictionary to store the percentage of employees who do not need training per department
efficiency = {}

# Calculate the percentage of employees who do not need training in each department
for department in departments:
    # Count of employees who do not need training in the department
    no_training_count = len(df[(df['Training'] == 'No') & (df['Department'] == department)])
    # Total employees in the department
    total_department_count = len(df[df['Department'] == department])
    # Calculate percentage
    percentage_no_training = (no_training_count / total_department_count) * 100
    efficiency[department] = percentage_no_training
    print(f"{department}: {percentage_no_training}% employees do not need training")

# Find the department with the highest percentage of employees who do not need training
most_efficient_department = max(efficiency, key=efficiency.get)

print("The most efficient department is:", most_efficient_department)
````

# Groupby and Aggregate Functions

Grouping and aggregation are two of the most frequently used operations in data analysis, especially while performing exploratory data analysis (EDA), where it is common to compare summary statistics across groups of data.


As an example, in the weather time-series data that you are working with, you may want to compare the average rainfall of various regions or compare temperature across different locations.



A grouping analysis can be thought of as having the following three parts:

*  Splitting the data into groups (e.g., groups of location, year, and month)
*  Applying a function on each group (e.g., mean, max, and min)
*  Combining the results into a data structure showing summary statistics

Sometimes feeling cold is more than about low temperatures; a windy day can also make you cold. A factor called the chill factor can be used to quantify the cold based on the wind speed and the temperature. The formula for the chill factor is given by


$ WCI = (10 * \sqrt{v} - v + 10.5) .(33 - T_{m}) $

v is the speed of the wind and $ T_{m} $ is the minimum temperature

Add a column for WCI and find the month with the lowest WCI.

````python
from math import sqrt
def wci(x): # x is a row of dataframe
    velocity = x['WindGustSpeed']
    minTemp = x['MinTemp']
    return ((10 * sqrt(velocity) - velocity + 10.5)*(33-minTemp))
````

````python
data['WCI'] = data.apply(wci,axis=1)
# axis=1 refers to operations performed along the row axis. When you use axis=1, the function is applied to each row individually.
````

Types of Joins

INNER JOIN (how='inner')

* Combines rows from both DataFrames where the join keys are matching in both.
* Example: If you have two DataFrames, one with employee details and another with department IDs, an inner join will return only those records where employee department IDs match the department IDs listed in the second DataFrame.

LEFT JOIN (how='left')

* Includes all rows from the left DataFrame and the matched rows from the right DataFrame.
* If there is no match, the result is NaN in the columns of the right DataFrame.
* Example: If you're merging employee records (left) with their department names (right), a left join ensures that all employees are listed, even if their departments are not in the second DataFrame.

RIGHT JOIN (how='right')

* Includes all rows from the right DataFrame and the matched rows from the left DataFrame.
* If there is no match, the result is NaN in the columns of the left DataFrame.
* Example: Using the same employee and department example, a right join ensures that all departments are listed, even if there are no employees associated with them in the left DataFrame.

FULL JOIN (how='outer')

* Combines all rows from both DataFrames where there is a match in either the left or right DataFrame.
* Where there is no match, the result will have NaN in the place of missing values.
* Example: A full join of the employee and department DataFrames would list all employees and all departments, filling in missing values with NaN where there is no match.

3. Merging and Appending DataFrames

Combining DataFrames using .merge() and .concat().

````python
# Merging two DataFrames on a key
df1 = pd.DataFrame({'key': ['K0', 'K1', 'K2'], 'A': ['A0', 'A1', 'A2']})
df2 = pd.DataFrame({'key': ['K0', 'K1', 'K2'], 'B': ['B0', 'B1', 'B2']})
merged = pd.merge(df1, df2, on='key')
merged
````

````python
# Concatenating DataFrames
concatenated = pd.concat([df1, df2], axis=0)
concatenated
````
